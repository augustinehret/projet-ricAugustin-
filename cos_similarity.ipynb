```python
import numpy as np


#on fait une fonction de tri (ici tri_fusion) pour pouvoir ensuite renvoeyr les films les plus proches, triés donc)

def fusion(l1, l2):

    if len(l1) == 0: return l2
    if len(l2) == 0: return l1
    

    if l1[0][1] > l2[0][1]:
        return [l1[0]] + fusion(l1[1:], l2)
    else:
        return [l2[0]] + fusion(l1, l2[1:])

def tri_fusion(l):

    if len(l) <= 1:
        return l
    
    milieu = len(l) // 2
    gauche = l[:milieu]
    droite = l[milieu:]

    return fusion(tri_fusion(gauche), tri_fusion(droite))





def cosine_similarity(i, Q, n):

    #on extrait le vecteur de l'ojbet i qu'on veut recommander 
    objet = Q[i]

    norme_objet = np.linalg.norm(objet)
    
    scores = []
    
    # on calcul l'angle de notre film avec les autres 

    for j in range(len(Q)):

        if j == i:
            continue
            
        v_j= Q[j]
        norme_j = np.linalg.norm(v_j)
        
        # On ajoute epsilon = 1e-9 pour éviter de diviser par 0 
        
        produit_scalaire = np.dot(objet, v_j)
        similitude = produit_scalaire / (norme_objet * norme_j + 1e-9)
        
        scores.append((j, similitude))
    
    #tri des scores (décroissant)
    scores_triés = tri_fusion(scores)
    
    return scores_triés[:n]




Q_test = np.array([
    [4.0, 1.0],  # Film 0 : Grosse Action (Blockbuster)
    [0.8, 0.2],  # Film 1 : Petite Action (Indé) -> Doit être proche du 0
    [1.0, 4.0]   # Film 2 : Très Romance
])

recos = cosine_similarity(0, Q_test, 2)

print("Films proches du Film 0 (Action) :")
for id_film, score in recos:
    print(f"- Film {id_film} avec un score de {score:.4f}")
```
